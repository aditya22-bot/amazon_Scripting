```python
from bs4 import BeautifulSoup
import requests
 
# Function to extract Product Title
def get_title(soup):
    try:
        # Outer Tag Object
        title = soup.find("span", attrs={"id":'productTitle'})
        title_value = title.string
        title_string = title_value.strip()
    except AttributeError:
        title_string = ""   
 
    return title_string
 
# Function to extract Product Price
def get_price(soup):
 
    try:
        price = soup.find("span", attrs={'class':'a-offscreen'}).string.strip()
 
    except AttributeError:
        price = ""  
 
    return price
def get_imgurl(soup):
    list_link=[]
    try:
        link1=soup.find_all("img",attrs={'id':"landingImage"})
        for i in link1:
            list_link.append(i.get('src'))
    except ArithmeticError:
        list_link=""
    return list_link
def get_details(soup):
    text=""
    try:
        link11=soup.find_all("tr",attrs={'id':"vatMessage"})
        for i in link11:
            text=i.text
    except ArithmeticError:
        text=""
    return text
         
if __name__ == '__main__':
    class SetEncoder(json.JSONEncoder):
        def default(self, obj):
            if isinstance(obj, set):
                return list(obj)
            return json.JSONEncoder.default(self, obj)
    res=[]
    headers={"Content-Type":"text" , "User-Agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36' , "Accept": '/' , 'Accept-Encoding' : 'gzip, deflate, br' , 'Connection': 'keep-alive' , 'cookie' : 'session-id=260-5103315-4097728; i18n-prefs=EUR; lc-acbde=en_GB; ubid-acbde=259-1485518-2684333; session-token=F+tK1/mNULvG3m/n0QKc7OerGSsLByJn4is1bN5hAiUMTV4LmQpG0i1FLJ1kmOgderfp/RhkrKXoQxk7UAZgNIACymBImgwlaSVcRYEHdcJB2r2J/h573yjbF99m0A4yTndqIhSGuWj6U5fFoogF/Y78N068IZhv/j8pwGTjACz7BZ90mvepln7QlTi9gVs3; csm-hit=tb:s-FEJ6V31R0AKJD1BFM69H|1644642128934&t:1644642130889&adb:adblk_no; session-id-time=2082754801l',}
    try:
        for i in link:
            url=i
#             soup = BeautifulSoup(req.content, "lxml")    
            webpage = requests.get(url, headers=headers)
            if webpage.status_code==404:
                raise Exception(i)
            else:
                soup = BeautifulSoup(webpage.content, "lxml")
                s=["Product Title","Product Price","Product url","Product Details"]
                res.append({s[0]:get_title(soup),s[1]:get_price(soup),s[2]:get_imgurl(soup),s[3]:get_details(soup)})   
    except Exception as e:
            res.append(e)
            
    with open("mydata.json", "w") as final:
        json.dump(res, final,cls=SetEncoder)
    
```
